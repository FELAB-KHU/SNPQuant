```python
import pandas as pd

data_path = r'../Data/data/'

# Read in the CSV file
crsp_data = pd.read_parquet(data_path + r'market_Top24_CRSP_Q1filter_Datedrop_fill.parquet', engine='fastparquet')
base_data = pd.read_parquet(data_path + r'market_Top24_FullPricingDate.parquet', engine='fastparquet')
```

```python
# crsp data에 base data를 merge
crsp_data['companyId'] = crsp_data['PERMNO']

# crsp date를 datatime으로 변환
crsp_data['pricingDate'] = pd.to_datetime(crsp_data['date'], format='%Y-%m-%d')
```

```python
crsp_data2 = pd.merge(crsp_data, base_data, on=['companyId', 'pricingDate'], how='left')
crsp_data2.head(5)
```

```python
crsp_data3 = crsp_data2.drop(crsp_data2.columns[14:], axis=1)
crsp_data3.head(5)
```

```python
# pricingDate, tickerSymbol	exchangeSymbol만 남기고 드롭
crsp_data4 = crsp_data3.drop(crsp_data3.columns[1:3], axis=1)
crsp_data5 = crsp_data4.drop(crsp_data4.columns[1:9], axis=1)
```

```python
# crsp_data5 유니크한 값만 남기기
crsp_data6 = crsp_data5.drop_duplicates()

# NaN 값 드롭
crsp_data6 = crsp_data6.dropna()
crsp_data6.reset_index(drop=True, inplace=True)
crsp_data6.to_csv(data_path + "ticker-companyId.csv", index=False)
```

```python
crsp_data6.head()
```

```python
oos1 = pd.read_csv(r"../PriceTrend\WORK_SPACE\new_model_res\20_5\20_5\ensem_res\ensem5_res_2022.csv", index_col=False)
oos2 = pd.read_csv(r"../PriceTrend\WORK_SPACE\new_model_res\20_5\20_5\ensem_res\ensem5_res_2023.csv", index_col=False)
```

```python
# oos1 과 oos2를 합침
oos = pd.concat([oos1, oos2])
oos
```

```python
oos.reset_index(drop=True, inplace=True)
oos = oos.drop(oos.columns[0], axis=1)
oos
```

```python
crsp_data6.rename(columns={'PERMNO': 'StockID'}, inplace=True)
```

```python
# oos 와 crsp_data6를 StockID 기준으로 합침
oos_crsp = pd.merge(oos, crsp_data6, on=['StockID'], how='left')
oos_crsp['ending_date'] = pd.to_datetime(oos_crsp['ending_date'])
```

```python
reddit = pd.read_csv(r"../reddit\reddit_factor_scaled.csv", index_col=0)
reddit.reset_index(inplace=True)
reddit.rename(columns={'factor': 'reddit_factor'}, inplace=True)
reddit.rename(columns={'index': 'ending_date'}, inplace=True)
reddit['ending_date'] = pd.to_datetime(reddit['ending_date'])
```

```python
# oos_crsp와 reddit을 ending_date 기준으로 합침
oos_crsp_reddit = pd.merge(oos_crsp, reddit, on=['ending_date'], how='left')
```

```python
oos_crsp_reddit.sort_values(by=['ending_date'], inplace=True)
```

```python
ohlcv = pd.read_feather(r'../PriceTrend\WORK_SPACE\data\processed_data\us_ret.feather')

ohlcv.drop(ohlcv.columns[10:], axis=1, inplace=True)
# 2022년 1월 1일부터 2023년 12월 31일까지의 데이터만 추출
ohlcv = ohlcv[ohlcv['Date'] >= '2022-01-01']
ohlcv.to_csv(r"../Data/data/final_ohlcv.csv", index=False)
```

```python
youtube = pd.read_csv(r"../Data/data/youtube_sentiment.csv", index_col=0)
youtube.reset_index(inplace=True)
youtube.rename(columns={'date': 'ending_date'}, inplace=True)
youtube['ending_date'] = pd.to_datetime(youtube['ending_date'], format='%Y-%m')
youtube.drop(youtube.columns[1:3], axis=1, inplace=True)
```

```python
# 투자 의사결정 lagging
youtube['positive'] = youtube['positive'].shift(-1)
youtube
```

```python
# 2017년부터 2023년의 모든 날짜를 생성
date = pd.date_range(start='2017-01-01', end='2023-12-31', freq='d')
date = pd.DataFrame(date, columns=['ending_date'])

# date와 youtube를 ending_date 기준으로 합침
youtube = pd.merge(date, youtube, on=['ending_date'], how='left')
youtube.ffill(inplace=True)
youtube
```

```python
# oos_crsp와 reddit을 ending_date 기준으로 합침
oos_crsp_reddit_youtube = pd.merge(oos_crsp_reddit, youtube, on=['ending_date'], how='left')
```

```python
oos_crsp_reddit_youtube['StockID'] = oos_crsp_reddit_youtube['StockID'].astype(str)
oos_crsp_reddit_youtube.to_csv(r"../Data/data/fianl_factor.csv", index=False)
oos_crsp_reddit_youtube.to_parquet(r"../Data/data/fianl_factor.parquet", index=False)
```
